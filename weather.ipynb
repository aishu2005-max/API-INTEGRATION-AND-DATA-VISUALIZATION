```python
import requests
import pandas as pd
```

```python
API_KEY = '94bdd9dabebed4caa9b3ee537ff59f30'
CITY = 'London'
URL = f'https://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}&units=metric'
```

```python
response = requests.get(URL)
```

```python
data = response.json()
```

```python
if response.status_code == 200:
    weather_data = {
        'City':data['name'],
        'Temperature (°C)':data['main']['temp'],
        'Humidity (%)':data['main']['humidity'],
        'weather':data['weather'][0]['description']
    }
    print(weather_data)
else:
    print("Error:", data['message'])
    
        
        
        
```

```python
df = pd.DataFrame([weather_data])
print(df)
print(df.head())
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
# plotting temperature
plt.figure(figsize=(8, 5))
sns.barplot(x='City', y='Temperature (°C)', hue='City', data=df,palette='Blues', legend=False, width=0.5)
plt.title('Temperature in °C')
plt.xlabel('City')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
#plotting humidity
plt.figure(figsize=(8, 5))
sns.barplot(x='City', y='Humidity (%)', hue='City', data=df, palette='Greens', legend=False )
plt.title('Humidity Percentage')
plt.xlabel('City')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
#plotting wind speed
plt.figure(figsize=(8, 5))
sns.barplot(x='City', y='weather', hue='City', data=df, palette='Oranges', legend=False,width=0.5)
plt.title('wind speed in m/s')
plt.xlabel('City')
plt.ylabel('weather')
plt.show()
```
